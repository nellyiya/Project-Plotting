Project Plotting - Matplotlib Seaborn, Histograms and Boxplots

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

```

load the dataset and print 15 rows


```python
fuel_econ= pd.read_csv('fuel_econ.csv')
fuel_econ.head(n=11)
```

Plot a histogram and a heatmap


```python
plt.figure(figsize=(10,6))
plt.hist(fuel_econ['co2'],bins=20, color='green', edgecolor='black')
plt.title('Histogram of CO2 Emissions')
plt.grid(True)
plt.show()
```

```python
# Selecting numeric columns
numeric_columns = fuel_econ.select_dtypes(include=['float64', 'int64'])

# Calculating the correlation matrix
corr_matrix = numeric_columns.corr()

# Plotting the heatmap
plt.figure(figsize=(12, 10))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Heatmap of Fuel Economy Data')
plt.show()
```
